```python
def mnk(x_,y_):
    n = len(x_)
    a = (n * np.dot(x_, y_) - np.sum(x_) * np.sum(y_)) / (n * np.sum(x_ ** 2) - np.sum(x_) ** 2)
    b = (np.sum(y_) - a * np.sum(x_)) / n
    return a,b
```
