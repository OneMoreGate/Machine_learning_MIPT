```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
df = pd.read_csv('tambov.csv', sep=';', comment='#', index_col=False, low_memory=False)
df.rename(columns={f'{df.columns[0]}' : 'Time'}, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['Time'] = pd.Series([dt.datetime.strptime(i, '%d.%m.%Y %H:%M').date() for i in df['Time']])
```

```python
sub_Time_U = pd.DataFrame({'Date': df['Time'], 'Humidity' : df['U']}).dropna()
sub_Time_T = pd.DataFrame({'Date': df['Time'], 'Temperature' : df['T']}).dropna()
```

```python
plt.plot(sub_Time_U['Date'], sub_Time_U['Humidity'], linewidth = 0.1)
```

```python
plt.plot(sub_Time_T['Date'], sub_Time_T['Temperature'], linewidth = 0.1)
```

```python
sub = df.loc[:,['T', 'Po']][df['DD'] == df['DD'].unique()[7]].dropna()
plt.scatter(sub['Po'], sub['T'], s=1)
plt.title(df['DD'].unique()[7])
```

```python
pd.pivot_table(df,
    values='T',
    index= 'DD',
    columns='Cm',
    aggfunc=np.mean
)
```

```python
uniq_days = df['Time'].unique()
mean_d = np.array([df['T'][df['Time'] == i].dropna().mean() for i in uniq_days])
day_mean = pd.DataFrame({'Date': uniq_days, 'Temp': mean_d})
```

```python
plt.plot(day_mean['Date'], day_mean['Temp'], linewidth = 0.4)
```

```python
year = pd.Series([i.year for i in df['Time']])
month = pd.Series([i.month for i in df['Time']])
day = pd.Series([i.day for i in df['Time']])
temp = df['T']

qwerty = pd.DataFrame({'year': year, 'month': month, 'day': day, 'temp': temp}).dropna()
```

```python
mean = np.array(pd.pivot_table(qwerty,
    values='temp',
    index= ['month', 'day'],
    aggfunc=np.mean
))
min = np.array(pd.pivot_table(qwerty,
    values='temp',
    index= ['month', 'day'],
    aggfunc=np.min
))
max = np.array(pd.pivot_table(qwerty,
    values='temp',
    index= ['month', 'day'],
    aggfunc=np.max
))

uniq_days_year = pd.Series([i for i in uniq_days if i.year == 2020 ]).unique()

plt.plot(uniq_days_year, mean[::-1])
plt.plot(uniq_days_year, min[::-1])
plt.plot(uniq_days_year, max[::-1])
```

```python
mean = (pd.pivot_table(qwerty,
    values='temp',
    index= ['year','month', 'day'],
    aggfunc=np.mean
))
unique_days = df['Time'].unique()

plt.plot(unique_days, mean[::-1], linewidth = 0.4)
```
